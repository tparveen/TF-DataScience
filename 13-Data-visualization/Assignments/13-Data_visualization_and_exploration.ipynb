```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Understanding the data: 


Region: String
Region the country belongs to


Family
The extent to which Family contributes to the calculation of the Happiness Score

Health (Life Expectancy): Numeric
The extent to which Life expectancy contributed to the calculation of the Happiness Score

Freedom : Numeric
The extent to which Freedom contributed to the calculation of the Happiness Score

Trust (Government Corruption): Numeric
The extent to which Perception of Corruption contributes to Happiness Score

Generosity: Numeric
The extent to which Generosity contributed to the calculation of the Happiness Score

Dystopia Residual: Numeric
The extent to which Dystopia Residual contributed to the calculation of the Happiness Score. (Dystopia is an imaginary country where most unhappy people live. The purpose in establishing Dystopia is to have a benchmark against which all countries can be favorably compared (no country performs more poorly than Dystopia) in terms of each of the six key variables, thus allowing each sub-bar to be of positive width. “Dystopia” was created in contrast to Utopia.)

The extent to which Dystopia Residual contributed to the calculation of the Happiness Score.

```python
# https://seaborn.pydata.org/introduction.html

# Choose one variable and plot that variable four different ways.
# choice of variable is Happiness score

# histogram of happiness score

plt.hist(data2['Happiness Score'], color='red', alpha=.5,label='Happiness Score')  # alpha just controls the opacity
plt.title('Happiness Score')
plt.show()
# happiness score - create 4 different plots like box, histogram, 
# hue in seaborn 

# region bar plot, 
```

```python
# barplot
# Setting the overall aesthetic.
sns.set(style="darkgrid")

# Use catplot (previous versions of seaborn use factorplot)
g = sns.catplot(x="Region", y="Happiness Score", data=data2,
                   height=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Tips")
#g.set_xlabels("")
#g.set_xlabels("")
plt.xticks(rotation=80) #'vertical'
plt.title('Barplot: Meal Tips by Time and Gender')
plt.show()
```
