####  For which might tossing out the records entirely make sense?

There are some variables whose values are more than 50% missing. Those values should be tossed out. Filling that much missing values will throw our results off.

AVG_MATH_4_SCORE                64.075067
AVG_MATH_8_SCORE                64.343164

```python
# dealing with missing values without taking time into consideration

#make a copy of the original df
df_education2 = df.copy()

#make a list of the columns that has missing values and we want to keep them
list_to_keep = ['ENROLL', 'TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE',                   
                'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE',    
                'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G',                      
                'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 
                'GRADES_9_12_G', 'GRADES_ALL_G']

for col in list_to_keep:
    df_education2.loc[:,col] = df_education2.loc[:,col].fillna(df_education2.loc[:,col].mean())
```

```python
# checking to see if the null values are filled

df_education2.isnull().sum()*100/df_education2.isnull().count() 
```

#### * 3-Now, take into account the time factor. Replicate your second answer but this time fill in the missing values by using a statistic that is calculated within the year of the observation. For example, if you want to fill a missing value for a variable with the mean of that variable, calculate the mean by using only the observations for that specific year.



```python
df_education3 = df.copy()

years = df_education3['YEAR'].unique()
print(years)

#years_nunique = df_education3['YEAR'].nunique()
#print(years_nunique)
#26
```

```python
for col in list_to_keep:
    for year in years:
        df_education3.loc[df_education3['YEAR'] == year, col] = df_education3.loc[df_education3['YEAR'] == year, 
                                                    col].fillna(df_education3.loc[df_education3['YEAR'] == year, col].mean())
```

```python
# checking to see if the null values are filled

df_education3.isnull().sum()*100/df_education3.isnull().count() 
```

#### * 4- This time, fill in the missing values using interpolation (extrapolation).


```python
df_education4 = df.copy()

for col in list_to_keep:
    df_education4.loc[:, col] = df_education4.loc[:, col].interpolate()
```

```python
# checking to see if the null values are filled

df_education4.isnull().sum()*100/df_education4.isnull().count()
```

#### * 5- Compare your results for the 2nd, 3rd, and 4th questions. Do you find any meaningful differences?

```python
fig = plt.figure(figsize=(20,100))

#consider the null columns that we filled by various methods

for i, col in enumerate(list_to_keep):
    df_plt = pd.concat([df[col], df_education2[col],
                     df_education3[col], df_education4[col]], axis=1)
    
    plt.subplot(len(list_to_keep), 4, i+1)
    plt.bar(["original", "overall mean", "year-by-year mean", "interpolation"], df_plt.median())
    plt.xticks(rotation='vertical')
    plt.title("Median of {}".format(col))
    
plt.tight_layout()
plt.show()
```

#### Results
If we look at the above plots, the plots shows the medium of all the cols that are in the list_to_keep list. So we plotted the median for all of those columns. The 3 columns (2,3,4) have their null values taken care of by various method. 
When we compare each of those medians with the first column, we can see that the year-by-year median is the most close to the 1st column which is the original without null values taken care of. So, year by year method of taking care of the null probably is the most accurate one. 

Interpolate looks to be the most off one.
